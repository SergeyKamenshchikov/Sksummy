CDFs:

```
def get_ner_spacy(text, num_of_sents, labels):
  final_entities_list=[]

  sentences_list=sent_tokenize(text)
  
  num_of_contexts=len(sentences_list)//num_of_sents
  if num_of_contexts==0:
    num_of_contexts=1

  contexts=numpy.array_split(sentences_list,num_of_contexts)
  contexts=[' '.join(context) for context in contexts]

  for context in contexts:
    doc=spacy_ner(context)
    for ent in doc.ents:
      if ent.label_ in labels:
        final_entities_list.append(ent.text)

  return(final_entities_list)
  
```
